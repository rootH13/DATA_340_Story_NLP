```python
import string
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
import nltk

def expand_pos_tag(tag):
    tag_definitions = {
         "NN": "Noun, singular or mass",
        "NNS": "Noun, plural",
        "NNP": "Proper noun, singular",
        "NNPS": "Proper noun, plural",
        "VB": "Verb, base form",
        "VBD": "Verb, past tense",
        "VBG": "Verb, gerund or present participle",
        "VBN": "Verb, past participle",
        "VBP": "Verb, non-3rd person singular present",
        "VBZ": "Verb, 3rd person singular present",
        "JJ": "Adjective",
        "JJR": "Adjective, comparative",
        "JJS": "Adjective, superlative",
        "RB": "Adverb",
        "RBR": "Adverb, comparative",
        "RBS": "Adverb, superlative",
        "PRP": "Personal pronoun",
        "PRP$": "Possessive pronoun",
        "IN": "Preposition or subordinating conjunction",
        "DT": "Determiner",
        "CC": "Coordinating conjunction",
        "CD": "Cardinal number",
        "MD": "Modal",
        "EX": "Existential there",
        "FW": "Foreign word",
        "POS": "Possessive ending",
        "RP": "Particle",
        "TO": "to",
        "WDT": "Wh-determiner",
        "WP": "Wh-pronoun",
        "WP$": "Possessive wh-pronoun",
        "WRB": "Wh-adverb"
    }
    return tag_definitions.get(tag, "Other")

def process_text(text1, text2):
    word_freq = Counter()
    all_tags = Counter()
    stop_words = set(stopwords.words('english'))

    # Combine the two texts
    combined_text = text1 + " " + text2

    # Convert to lower case and remove punctuation, apostrophes, and quotations
    combined_text = combined_text.lower().translate(str.maketrans('', '', string.punctuation + "‘’“”"))

    # Tokenize and filter out stopwords
    tokens = word_tokenize(combined_text)
    filtered_tokens = [word for word in tokens if word not in stop_words]
    word_freq.update(filtered_tokens)

    # POS Tagging and expansion
    tagged = pos_tag(filtered_tokens)
    all_tags.update([expand_pos_tag(tag) for _, tag in tagged])

    return word_freq, all_tags

# Sample text blurbs
text_blurb1 = """Title: The Mysterious Disappearance of Mr. Smith

It was a chilly autumn evening when Mr. John Smith, a respected member of the community, arrived at his estate in a state of distress. He had been away on a business trip and had returned to find that his beloved wife, Mrs. Emily Smith, was nowhere to be found.

"Good heavens, John!" exclaimed Mrs. Nancy Brown, the Smiths' next-door neighbor. "What has happened? Where is your dear Emily?"

Mr. Smith's face was pale and worried as he replied, "I don't know, Nancy. She was here just a short while ago, but now she's gone without a trace."

Mrs. Brown's eyes widened with concern as she asked, "Do you have any idea where she might have gone or who might have taken her?"

Mr. Smith shook his head mutely, his brow furrowed in worry. After a moment of silence, he spoke up, "I fear the worst, Nancy. I fear that something terrible has happened to my dear Emily."

Just then, there was a knock at the door. It was Inspector James, the local law enforcement officer. "Evening, Mr. Smith," he said with a somber tone. "I heard about your wife's disappearance and wanted to offer my assistance."

Mr. Smith thanked the inspector gratefully and told him everything he knew about the situation. Inspector James listened intently before saying, "Don't worry, sir. We will do our best to find your wife and bring her home safely."

As the night wore on, the inspector questioned Mr. Smith further and searched the estate for clues. But despite their efforts, no trace of Mrs. Smith could be found.

One day passed, and then another, but still no sign of Mrs. Smith. Mr. Smith grew more and more desperate, feeling as though he would never see his beloved wife again. But he refused to give up hope, knowing that she was out there somewhere, waiting to be found.

And so, the search continued, with Inspector James leading the way. They followed every lead, no matter how small, until finally, they received a tip that led them to an unexpected discovery.

As it turned out, Mrs. Smith had not disappeared without a reason. She had left on a secret mission, determined to uncover a hidden treasure that only she knew existed. And after weeks of searching, she finally found what she was looking for – a chest filled with gold coins and precious jewels.

The news of Mrs. Smith's successful adventure spread quickly throughout the town, and soon everyone was talking about it. Even Mr. Smith couldn't help but feel a sense of pride and admiration for his wife's bravery and determination.

In the end, Mrs. Smith returned home to her husband, her chest overflowing with riches, and her heart full of joy. And although the journey had been long and difficult, she knew that it had all been worth it in the end."""
text_blurb2 = """Title: The Mysterious Disappearance of the Family Heirloom

One dark and stormy night, Mr. and Mrs. Smith sat in their cozy study, deep in conversation about their recently stolen family heirloom, a priceless diamond necklace that had been passed down for generations.

"Well, I can't believe someone would do this to us," Mrs. Smith said, her voice trembling with anger.

"It's unfortunate, my dear," Mr. Smith replied, his brow furrowed in concern. "But we must not let this theft get the best of us."

Just then, there was a knock at the door. Mrs. Smith got up to answer it, expecting it to be the local police officer with whom she had made arrangements to report the crime. But upon opening the door, she found herself face to face with an unexpected visitor - a tall, slender man with piercing blue eyes and a mischievous grin.

"Can I help you?" Mrs. Smith asked, trying to hide her unease.

"Oh, good evening, Mrs. Smith," the stranger replied, tipping his hat. "My name is Jack, and I might have some information regarding your missing necklace."

Mrs. Smith's curiosity piqued, she invited Jack inside to discuss further. As they chatted, Jack revealed himself to be a skilled detective who had been tracking the thief for weeks.

"I think I know where your necklace might be," Jack said, his eyes sparkling with excitement. "Would you like me to take a look around?"

Mr. Smith nodded eagerly, hoping against hope that they might soon recover their beloved family heirloom. And so, Jack set off on his mission to solve the mystery of the disappearing necklace. Little did anyone realize, Jack was about to uncover more than just the location of the stolen jewels..."""
word_freq, all_tags = process_text(text_blurb1, text_blurb2)

# Print most common words and POS tags
print("Most common words:")
for word, freq in word_freq.most_common(10):
    print(f"{word}: {freq}")

print("\nMost common Part Of Speech tags:")
for tag, count in all_tags.most_common(10):
    print(f"{tag}: {count}")
```

```python
import matplotlib.pyplot as plt
import spacy

# Load the SpaCy model
nlp = spacy.load('en_core_web_sm')

def analyze_sentence_structure(text1, text2):
    sentence_count = 0
    word_count = 0
    pos_count = {}

    # Process each text separately and combine the results
    for text in [text1, text2]:
        doc = nlp(text)
        for sent in doc.sents:
            sentence_count += 1
            for token in sent:
                word_count += 1
                pos_count[token.pos_] = pos_count.get(token.pos_, 0) + 1

    return sentence_count, word_count, pos_count

def plot_pos_distribution(pos_count):
    labels, values = zip(*pos_count.items())
    plt.figure(figsize=(10, 5))
    plt.bar(labels, values)
    plt.title('Part Of Speech Distribution')
    plt.xlabel('POS Tags')
    plt.ylabel('Frequency')
    plt.xticks(rotation=90)  # Rotate labels for better readability
    plt.show()

# Sample text blurbs
text_blurb1 = """Title: The Mysterious Disappearance of Mr. Smith

It was a chilly autumn evening when Mr. John Smith, a respected member of the community, arrived at his estate in a state of distress. He had been away on a business trip and had returned to find that his beloved wife, Mrs. Emily Smith, was nowhere to be found.

"Good heavens, John!" exclaimed Mrs. Nancy Brown, the Smiths' next-door neighbor. "What has happened? Where is your dear Emily?"

Mr. Smith's face was pale and worried as he replied, "I don't know, Nancy. She was here just a short while ago, but now she's gone without a trace."

Mrs. Brown's eyes widened with concern as she asked, "Do you have any idea where she might have gone or who might have taken her?"

Mr. Smith shook his head mutely, his brow furrowed in worry. After a moment of silence, he spoke up, "I fear the worst, Nancy. I fear that something terrible has happened to my dear Emily."

Just then, there was a knock at the door. It was Inspector James, the local law enforcement officer. "Evening, Mr. Smith," he said with a somber tone. "I heard about your wife's disappearance and wanted to offer my assistance."

Mr. Smith thanked the inspector gratefully and told him everything he knew about the situation. Inspector James listened intently before saying, "Don't worry, sir. We will do our best to find your wife and bring her home safely."

As the night wore on, the inspector questioned Mr. Smith further and searched the estate for clues. But despite their efforts, no trace of Mrs. Smith could be found.

One day passed, and then another, but still no sign of Mrs. Smith. Mr. Smith grew more and more desperate, feeling as though he would never see his beloved wife again. But he refused to give up hope, knowing that she was out there somewhere, waiting to be found.

And so, the search continued, with Inspector James leading the way. They followed every lead, no matter how small, until finally, they received a tip that led them to an unexpected discovery.

As it turned out, Mrs. Smith had not disappeared without a reason. She had left on a secret mission, determined to uncover a hidden treasure that only she knew existed. And after weeks of searching, she finally found what she was looking for – a chest filled with gold coins and precious jewels.

The news of Mrs. Smith's successful adventure spread quickly throughout the town, and soon everyone was talking about it. Even Mr. Smith couldn't help but feel a sense of pride and admiration for his wife's bravery and determination.

In the end, Mrs. Smith returned home to her husband, her chest overflowing with riches, and her heart full of joy. And although the journey had been long and difficult, she knew that it had all been worth it in the end."""
text_blurb2 = """Title: The Mysterious Disappearance of the Family Heirloom

One dark and stormy night, Mr. and Mrs. Smith sat in their cozy study, deep in conversation about their recently stolen family heirloom, a priceless diamond necklace that had been passed down for generations.

"Well, I can't believe someone would do this to us," Mrs. Smith said, her voice trembling with anger.

"It's unfortunate, my dear," Mr. Smith replied, his brow furrowed in concern. "But we must not let this theft get the best of us."

Just then, there was a knock at the door. Mrs. Smith got up to answer it, expecting it to be the local police officer with whom she had made arrangements to report the crime. But upon opening the door, she found herself face to face with an unexpected visitor - a tall, slender man with piercing blue eyes and a mischievous grin.

"Can I help you?" Mrs. Smith asked, trying to hide her unease.

"Oh, good evening, Mrs. Smith," the stranger replied, tipping his hat. "My name is Jack, and I might have some information regarding your missing necklace."

Mrs. Smith's curiosity piqued, she invited Jack inside to discuss further. As they chatted, Jack revealed himself to be a skilled detective who had been tracking the thief for weeks.

"I think I know where your necklace might be," Jack said, his eyes sparkling with excitement. "Would you like me to take a look around?"

Mr. Smith nodded eagerly, hoping against hope that they might soon recover their beloved family heirloom. And so, Jack set off on his mission to solve the mystery of the disappearing necklace. Little did anyone realize, Jack was about to uncover more than just the location of the stolen jewels..."""
sentence_count, word_count, pos_count = analyze_sentence_structure(text_blurb1, text_blurb2)

print(f"Total number of sentences: {sentence_count}")
print(f"Total number of words: {word_count}")
print("Part Of Speech counts:", pos_count)

# Visualize the POS distribution
plot_pos_distribution(pos_count)
```

```python
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from textblob import TextBlob
import textstat

# Function to analyze the style of a given text
def analyze_style(text):
    sentences = sent_tokenize(text)
    if len(sentences) == 0:
        return 0, 0, 0  # Avoid division by zero

    average_sentence_length = sum(len(word_tokenize(sentence)) for sentence in sentences) / len(sentences)
    blob = TextBlob(text)
    passive_sentences = sum(1 for sentence in blob.sentences if sentence.tags and sentence.tags[0][1] in ['VBN', 'VBD'])

    flesch_reading_ease = textstat.flesch_reading_ease(text)

    return average_sentence_length, passive_sentences, flesch_reading_ease

# Sample text blurbs
text_blurb1 = """Title: The Mysterious Disappearance of Mr. Smith

It was a chilly autumn evening when Mr. John Smith, a respected member of the community, arrived at his estate in a state of distress. He had been away on a business trip and had returned to find that his beloved wife, Mrs. Emily Smith, was nowhere to be found.

"Good heavens, John!" exclaimed Mrs. Nancy Brown, the Smiths' next-door neighbor. "What has happened? Where is your dear Emily?"

Mr. Smith's face was pale and worried as he replied, "I don't know, Nancy. She was here just a short while ago, but now she's gone without a trace."

Mrs. Brown's eyes widened with concern as she asked, "Do you have any idea where she might have gone or who might have taken her?"

Mr. Smith shook his head mutely, his brow furrowed in worry. After a moment of silence, he spoke up, "I fear the worst, Nancy. I fear that something terrible has happened to my dear Emily."

Just then, there was a knock at the door. It was Inspector James, the local law enforcement officer. "Evening, Mr. Smith," he said with a somber tone. "I heard about your wife's disappearance and wanted to offer my assistance."

Mr. Smith thanked the inspector gratefully and told him everything he knew about the situation. Inspector James listened intently before saying, "Don't worry, sir. We will do our best to find your wife and bring her home safely."

As the night wore on, the inspector questioned Mr. Smith further and searched the estate for clues. But despite their efforts, no trace of Mrs. Smith could be found.

One day passed, and then another, but still no sign of Mrs. Smith. Mr. Smith grew more and more desperate, feeling as though he would never see his beloved wife again. But he refused to give up hope, knowing that she was out there somewhere, waiting to be found.

And so, the search continued, with Inspector James leading the way. They followed every lead, no matter how small, until finally, they received a tip that led them to an unexpected discovery.

As it turned out, Mrs. Smith had not disappeared without a reason. She had left on a secret mission, determined to uncover a hidden treasure that only she knew existed. And after weeks of searching, she finally found what she was looking for – a chest filled with gold coins and precious jewels.

The news of Mrs. Smith's successful adventure spread quickly throughout the town, and soon everyone was talking about it. Even Mr. Smith couldn't help but feel a sense of pride and admiration for his wife's bravery and determination.

In the end, Mrs. Smith returned home to her husband, her chest overflowing with riches, and her heart full of joy. And although the journey had been long and difficult, she knew that it had all been worth it in the end."""
text_blurb2 = """Title: The Mysterious Disappearance of the Family Heirloom

One dark and stormy night, Mr. and Mrs. Smith sat in their cozy study, deep in conversation about their recently stolen family heirloom, a priceless diamond necklace that had been passed down for generations.

"Well, I can't believe someone would do this to us," Mrs. Smith said, her voice trembling with anger.

"It's unfortunate, my dear," Mr. Smith replied, his brow furrowed in concern. "But we must not let this theft get the best of us."

Just then, there was a knock at the door. Mrs. Smith got up to answer it, expecting it to be the local police officer with whom she had made arrangements to report the crime. But upon opening the door, she found herself face to face with an unexpected visitor - a tall, slender man with piercing blue eyes and a mischievous grin.

"Can I help you?" Mrs. Smith asked, trying to hide her unease.

"Oh, good evening, Mrs. Smith," the stranger replied, tipping his hat. "My name is Jack, and I might have some information regarding your missing necklace."

Mrs. Smith's curiosity piqued, she invited Jack inside to discuss further. As they chatted, Jack revealed himself to be a skilled detective who had been tracking the thief for weeks.

"I think I know where your necklace might be," Jack said, his eyes sparkling with excitement. "Would you like me to take a look around?"

Mr. Smith nodded eagerly, hoping against hope that they might soon recover their beloved family heirloom. And so, Jack set off on his mission to solve the mystery of the disappearing necklace. Little did anyone realize, Jack was about to uncover more than just the location of the stolen jewels..."""
# Analyze style for each text blurb
style_data = [
    ("Text Blurb 1", *analyze_style(text_blurb1)),
    ("Text Blurb 2", *analyze_style(text_blurb2))
]

# Extracting data for visualization
blurbs, avg_lengths, passive_counts, reading_eases = zip(*style_data)

# Visualization
plt.figure(figsize=(18, 6))  # Increase the size of the figure

# Average Sentence Length
plt.subplot(1, 3, 1)
plt.bar(blurbs, avg_lengths, color='orchid')
plt.title('Average Sentence Length')
plt.xticks(rotation=45, fontsize=10)  # Rotate and adjust font size
plt.ylabel('Words per Sentence')

# Passive Sentences Count
plt.subplot(1, 3, 2)
plt.bar(blurbs, passive_counts, color='lightpink')
plt.title('Number of Passive Sentences')
plt.xticks(rotation=45, fontsize=8)  # Rotate and adjust font size
plt.ylabel('Count')

# Flesch Reading Ease Score
plt.subplot(1, 3, 3)
plt.bar(blurbs, reading_eases, color='tan')
plt.title('Flesch Reading Ease Score')
plt.xticks(rotation=45, fontsize=8)  # Rotate and adjust font size
plt.ylabel('Score')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
