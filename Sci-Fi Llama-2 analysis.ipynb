```python
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from textblob import TextBlob
import textstat

# Function to analyze the style of a given text
def analyze_style(text):
    sentences = sent_tokenize(text)
    if len(sentences) == 0:
        return 0, 0, 0  # Avoid division by zero

    average_sentence_length = sum(len(word_tokenize(sentence)) for sentence in sentences) / len(sentences)
    blob = TextBlob(text)
    passive_sentences = sum(1 for sentence in blob.sentences if sentence.tags and sentence.tags[0][1] in ['VBN', 'VBD'])

    flesch_reading_ease = textstat.flesch_reading_ease(text)

    return average_sentence_length, passive_sentences, flesch_reading_ease

# Sample text blurbs
text_blurb1 = """Title: The Lost Colony of Zorvath

Chapter 1 - Departure

Commander Valtor stood upon the bridge of the interstellar vessel, Zorvath's Hope. "Prepare for departure, crew," he barked. "We must reach our new homeworld before the xenomorph infestation spreads."

"Aye, Commander," replied Lieutenant Nyx, navigating through the ship's systems.

Valtor glared at the viewscreen, displaying a swirling vortex of light and color. "I only wish we could escape this wretched galaxy."

Nyx chuckled dryly. "Such sentiments are...unseemly, Commander."

Chapter 2 - Encounter

The ship hurtled towards an uncharted planet, shrouded in darkness. Suddenly, a strange energy beam pierced the hull, crippling critical systems.

Valtor cursed under his breath. "What sorcery is this?"

Nyx shook her head, her eyes fixed on the viewscreen. "I cannot discern its origin, Commander."

A voice echoed from the speakers. "Greetings, inhabitants of Zorvath's Hope. We mean you no harm, but your presence here is...inconvenient."

Valtor squinted at the screen. "Who dares address us thus? Identify yourself!"

Chapter 3 - Conflict

A group of beings materialized within the bridge. They wore strange attire, adorned with glowing symbols. One stepped forward, offering a hand in peace.

Lieutenant Fusek eyed the stranger warily. "And what business do you have here?"

The being smiled. "We seek to protect your kind from the impending doom that threatens all life in this galaxy."

Valtor crossed his arms, suspicious. "How can we trust your words?"

The being bowed. "You may not, but know this: we share your desire to find a habitable world. Together, we can achieve that goal."

Chapter 4 - Decision

Hours passed as the Zorvathians deliberated among themselves. Valtor leaned against a console, lost in thought. "This offers hope, yet I fear their intentions..."

Nyx nodded in agreement. "Yes, Commander. But consider: if they truly possess the knowledge we need, perhaps allying ourselves might be wise."

Fusek spoke up. "I say we take the chance. Our survival depends on it."

Valtor straightened, determination etched on his"""
text_blurb2 = """Title: The Last Hope

Chapter 1 - Discovery

Captain Jack "Hawk" Hawkins gazed upon the planet, his eyes squinting against the glow of the setting sun. Upon the bridge of his trusty old ship, the Starblaze, he gathered his senior officers for a briefing on their latest mission.

"Gentlemen," he began, "our scans have picked up an unusual energy signature emanating from this uncharted world. It could be anything – a derelict spacecraft, a natural phenomenon... or something more sinister."

Scientist Dr. Samantha Chen nodded thoughtfully beside him, her dark hair pulled back in a tight bun. "Our probes indicate a dense atmosphere and unusual geological formations. It's definitely worth exploring."

Commander James "Raven" Reyes leaned forward from the weapons station, his piercing blue eyes locked onto the captain. "What's our plan of attack, Captain? We can't just barge in there without any prep work."

Chapter 2 - Encounter

As they descended through the atmospheric layers, the crew detected strange readings on their instruments. Radar pinged off unknown objects, while the gravity sensors spiked with unexplained fluctuations. Captain Hawkins steeled himself as the landing craft touched down upon the surface.

The first rays of morning light illuminated the alien landscape outside. A hostile wind howled across the barren wasteland, whipping sand devils into frenzied dance. Jack emerged from the ship, boots crunching on the rough terrain as he led his team towards the mysterious structure.

Dr. Chen spoke softly to herself, examining data on her tablet. "It's definitely not organic in origin. Whatever it is, it's been here for millions of years at least."

Commander Reyes flexed his muscles, checking his weapon's settings. "I don't like this one bit. What if whatever made that thing isn't dead yet?"

Chapter 3 - Confrontation

Inching closer to the massive structure, they found themselves face-to-face with an unexpected guardian. An ancient being, seemingly carved from the same material as the structure itself, stood watchful before them. Its voice rumbled within their minds, its intentions uncertain.

"Why have you come here?" the guardian asked, sending a shiver down the spines of those present.

Captain Hawkins stepped forward, clearing his throat. "We mean no harm. Our only desire is to unravel your secrets"""
# Analyze style for each text blurb
style_data = [
    ("Text Blurb 1", *analyze_style(text_blurb1)),
    ("Text Blurb 2", *analyze_style(text_blurb2))
]

# Extracting data for visualization
blurbs, avg_lengths, passive_counts, reading_eases = zip(*style_data)

# Visualization
plt.figure(figsize=(18, 6))  # Increase the size of the figure

# Average Sentence Length
plt.subplot(1, 3, 1)
plt.bar(blurbs, avg_lengths, color='orchid')
plt.title('Average Sentence Length')
plt.xticks(rotation=45, fontsize=10)  # Rotate and adjust font size
plt.ylabel('Words per Sentence')

# Passive Sentences Count
plt.subplot(1, 3, 2)
plt.bar(blurbs, passive_counts, color='lightpink')
plt.title('Number of Passive Sentences')
plt.xticks(rotation=45, fontsize=8)  # Rotate and adjust font size
plt.ylabel('Count')

# Flesch Reading Ease Score
plt.subplot(1, 3, 3)
plt.bar(blurbs, reading_eases, color='tan')
plt.title('Flesch Reading Ease Score')
plt.xticks(rotation=45, fontsize=8)  # Rotate and adjust font size
plt.ylabel('Score')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
